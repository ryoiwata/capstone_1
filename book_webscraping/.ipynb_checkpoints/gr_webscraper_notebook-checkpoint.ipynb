```python
author_sub_soup = BeautifulSoup(author_sub_page.text, 'html.parser')
author_sub_soup
```

```python
language = book_sub_soup.find_all(itemprop="inLanguage")[0].text
language
```

```python
birth_place = ""
for char in author_sub_soup.find("div", class_ = "dataTitle").next_siblings:
    if char.name == "div":
        break
    else:
        birth_place += str(char)
birth_country = birth_place.strip().split(",")[-1]
birth_country = birth_country.strip("The ")
birth_country = birth_country.split("\n")[0]


birth_country
```
